```python
from pyvispoly import (
    Point,
    Polygon,
    PolygonWithHoles,
    VisibilityPolygonCalculator,
)
from pyvispoly.plotting import plot_polygon
```

```python
poly_a = PolygonWithHoles(Polygon([Point(0, 0), Point(3, 0), Point(1, 1), Point(0, 1)]))
poly_b = PolygonWithHoles(Polygon([Point(1, 0), Point(2, 0), Point(2, 1), Point(1, 1)]))
```

```python
float(poly_a.outer_boundary().area())
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
plot_polygon(poly_a, ax=ax, alpha=0.5, color="blue")
plot_polygon(poly_b, ax=ax, alpha=0.5, color="red")
plt.show()
```

```python
fig, ax = plt.subplots()
for poly in poly_a.difference(poly_b):
    plot_polygon(poly, ax=ax, alpha=0.5)
plt.show()
```

```python
u_poly = PolygonWithHoles(
    [
        Point(0, 0),
        Point(0, 3),
        Point(1, 3),
        Point(1, 1),
        Point(2, 1),
        Point(2, 3),
        Point(3, 3),
        Point(3, 0),
    ][::-1]
)
float(u_poly.outer_boundary().area())
```

```python
fig, ax = plt.subplots()
plot_polygon(u_poly, ax=ax, alpha=0.5, color="blue")
# equal aspect ratio
ax.set_aspect("equal", "box")
plt.show()
```

```python
vispoly = VisibilityPolygonCalculator(u_poly)
q = Point(0, 0.1)
visble = vispoly.compute_visibility_polygon(q)
```

```python
fig, ax = plt.subplots()
plot_polygon(
    u_poly, ax=ax, alpha=1.0, color="lightgrey", linewidth=1, edgecolor="black"
)
plot_polygon(visble, ax=ax, alpha=0.4, color="blue")
ax.plot(q.x(), q.y(), "o", color="black")
# equal aspect ratio
ax.set_aspect("equal", "box")
plt.show()
```

```python
q = Point(3, 0)
visble = vispoly.compute_visibility_polygon(q)
fig, ax = plt.subplots()
# ax.set_facecolor("lightgrey")
plot_polygon(
    u_poly, ax=ax, alpha=1.0, color="lightgrey", linewidth=1, edgecolor="black"
)
plot_polygon(visble, ax=ax, alpha=0.4, color="blue")
assert visble.contains(q)
assert visble.contains(Point(2, 0))
assert not visble.contains(Point(0, 3))
ax.plot(q.x(), q.y(), "o", color="black")
# equal aspect ratio
ax.set_aspect("equal", "box")
plt.show()
```

```python
q = Point(3, 0)
visble = vispoly.compute_visibility_polygon(q)
sample_points = visble.interior_sample_points()
fig, ax = plt.subplots()
# ax.set_facecolor("lightgrey")
plot_polygon(
    u_poly, ax=ax, alpha=1.0, color="lightgrey", linewidth=1, edgecolor="black"
)
plot_polygon(visble, ax=ax, alpha=0.4, color="blue")
assert visble.contains(q)
assert visble.contains(Point(2, 0))
assert not visble.contains(Point(0, 3))
ax.plot(q.x(), q.y(), "o", color="black")
ax.plot(
    [float(p.x()) for p in sample_points],
    [float(p.y()) for p in sample_points],
    "x",
    color="black",
)
# equal aspect ratio
ax.set_aspect("equal", "box")
plt.show()
```

```python
q = Point(3, 0)
visble = vispoly.compute_visibility_polygon(q)
remaining_parts = u_poly.difference(PolygonWithHoles(visble))
sample_points = sum((poly.interior_sample_points() for poly in remaining_parts), [])
fig, ax = plt.subplots()
# ax.set_facecolor("lightgrey")
plot_polygon(
    u_poly, ax=ax, alpha=1.0, color="lightgrey", linewidth=1, edgecolor="black"
)
plot_polygon(visble, ax=ax, alpha=0.4, color="blue")
assert visble.contains(q)
assert visble.contains(Point(2, 0))
assert not visble.contains(Point(0, 3))
ax.plot(q.x(), q.y(), "o", color="black")
ax.plot(
    [float(p.x()) for p in sample_points],
    [float(p.y()) for p in sample_points],
    "x",
    color="black",
)
# equal aspect ratio
ax.set_aspect("equal", "box")
plt.show()
```

```python
poly_a = PolygonWithHoles([Point(0, 0), Point(4, 0), Point(4, 4), Point(0, 4)])
poly_b = PolygonWithHoles([Point(1, 1), Point(3, 1), Point(3, 3), Point(1, 3)])
poly = poly_a.difference(poly_b)[0]
fig, ax = plt.subplots()
plot_polygon(poly, ax=ax, alpha=0.5, color="blue")
sample_points = poly.interior_sample_points()
ax.plot(
    [float(p.x()) for p in sample_points],
    [float(p.y()) for p in sample_points],
    "x",
    color="black",
)
plt.show()
```
