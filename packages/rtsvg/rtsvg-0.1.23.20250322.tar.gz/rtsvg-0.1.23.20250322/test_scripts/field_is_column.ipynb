```python
import pandas as pd
import numpy  as np
import sys
sys.path.insert(1, '../rtsvg')
from rtsvg import *
rt = RACETrack()
```

```python
df = pd.DataFrame({'field':['field'], '5':[5], '1':['1']})
rt.displaySVG(
 rt.tile([
  rt.tile([
      rt.histogram(df, bin_by='field',h=36), 
      rt.histogram(df, bin_by='5',    h=36), 
      rt.histogram(df, bin_by='1',    h=36)
 ]),
  rt.tile([
      rt.histogram(df, bin_by='field', count_by='field', h=36), 
      rt.histogram(df, bin_by='5',     count_by='field', h=36), 
      rt.histogram(df, bin_by='1',     count_by='field', h=36)
 ]),
  rt.tile([
      rt.histogram(df, bin_by='field', count_by='5', h=36), 
      rt.histogram(df, bin_by='5',     count_by='5', h=36), 
      rt.histogram(df, bin_by='1',     count_by='5', h=36)
 ]),
  rt.tile([
      rt.histogram(df, bin_by='field', count_by='1', h=36), 
      rt.histogram(df, bin_by='5',     count_by='1', h=36), 
      rt.histogram(df, bin_by='1',     count_by='1', h=36)
 ]),
  rt.tile([
      rt.histogram(df, bin_by='field',color_by='field', h=36), 
      rt.histogram(df, bin_by='5',    color_by='field', h=36), 
      rt.histogram(df, bin_by='1',    color_by='field', h=36)
 ]),
  rt.tile([
      rt.histogram(df, bin_by='field', count_by='field', color_by='field', h=36), 
      rt.histogram(df, bin_by='5',     count_by='field', color_by='field', h=36), 
      rt.histogram(df, bin_by='1',     count_by='field', color_by='field', h=36)
 ]),
  rt.tile([
      rt.histogram(df, bin_by='field', count_by='5', color_by='field', h=36), 
      rt.histogram(df, bin_by='5',     count_by='5', color_by='field', h=36), 
      rt.histogram(df, bin_by='1',     count_by='5', color_by='field', h=36)
 ]),
  rt.tile([
      rt.histogram(df, bin_by='field', count_by='1', color_by='field', h=36), 
      rt.histogram(df, bin_by='5',     count_by='1', color_by='field', h=36), 
      rt.histogram(df, bin_by='1',     count_by='1', color_by='field', h=36)
 ]),
  rt.tile([
      rt.histogram(df, bin_by='field',color_by='5', h=36), 
      rt.histogram(df, bin_by='5',    color_by='5', h=36), 
      rt.histogram(df, bin_by='1',    color_by='5', h=36)
 ]),
  rt.tile([
      rt.histogram(df, bin_by='field', count_by='field', color_by='5', h=36), 
      rt.histogram(df, bin_by='5',     count_by='field', color_by='5', h=36), 
      rt.histogram(df, bin_by='1',     count_by='field', color_by='5', h=36)
 ]),
  rt.tile([
      rt.histogram(df, bin_by='field', count_by='5', color_by='5', h=36), 
      rt.histogram(df, bin_by='5',     count_by='5', color_by='5', h=36), 
      rt.histogram(df, bin_by='1',     count_by='5', color_by='5', h=36)
 ]),
  rt.tile([
      rt.histogram(df, bin_by='field', count_by='1', color_by='5', h=36), 
      rt.histogram(df, bin_by='5',     count_by='1', color_by='5', h=36), 
      rt.histogram(df, bin_by='1',     count_by='1', color_by='5', h=36)
 ]),
  rt.tile([
      rt.histogram(df, bin_by='field',color_by='1', h=36), 
      rt.histogram(df, bin_by='5',    color_by='1', h=36), 
      rt.histogram(df, bin_by='1',    color_by='1', h=36)
 ]),
  rt.tile([
      rt.histogram(df, bin_by='field', count_by='field', color_by='1', h=36), 
      rt.histogram(df, bin_by='5',     count_by='field', color_by='1', h=36), 
      rt.histogram(df, bin_by='1',     count_by='field', color_by='1', h=36)
 ]),
  rt.tile([
      rt.histogram(df, bin_by='field', count_by='5', color_by='1', h=36), 
      rt.histogram(df, bin_by='5',     count_by='5', color_by='1', h=36), 
      rt.histogram(df, bin_by='1',     count_by='5', color_by='1', h=36)
 ]),
  rt.tile([
      rt.histogram(df, bin_by='field', count_by='1', color_by='1', h=36), 
      rt.histogram(df, bin_by='5',     count_by='1', color_by='1', h=36), 
      rt.histogram(df, bin_by='1',     count_by='1', color_by='1', h=36)
 ]),
],horz=False)
)
```
